```python
file=open('C:/Users/User/Desktop/tail20_diamond.txt','r')
txt1=''
for line in file:
    txt1 += line
diamond_tail=txt1.split("==>")

diamond_tail = diamond_tail[1:]

def trimming(x):
    split_list=x.split('<==')
    name=split_list[0].strip().split('_')[0]
    contents = split_list[1]
    contents_list=contents.split("\n")
    hit_tail=[]
    for i in contents_list:
        if i == '':
            pass
        else:
            hit_tail.append(i.split('\t')[0])
    return [name,hit_tail]

diamond_dict={}
for i in range(len(diamond_tail)):
    func_result=trimming(diamond_tail[i])
    diamond_dict[func_result[0]]=func_result[1]
    
diamond_dict_final={}
for i in diamond_dict.keys():
    diamond_dict_final[i]=diamond_dict[i][-1]
    
dia_df=pd.DataFrame(pd.Series(diamond_dict_final))
dia_df=dia_df.reset_index()
dia_df.rename(columns={'index': 'sra', 0: 'final_sequence_diamond'}, inplace=True)

#diamond 결과도 따로 저장하기
dia_df.to_csv("C:/Users/User/Desktop/profiling/diamond_tail.csv",index=False)
```
